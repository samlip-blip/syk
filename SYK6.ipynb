# Backgound

__Anatomy of a Black Hole__

A black hole has a few key parts:

- Singularity: This is the very center, where matter is infinitely dense. The laws of physics as we know them break down here.

- Event Horizon: The "point of no return." If something crosses this invisible boundary, it’s doomed to fall into the black hole.

- Accretion Disk: Some black holes are surrounded by swirling hot gas and dust, which glow brightly and help scientists detect them.

__Types of Black Holes__

There are different sizes of black holes:

- Stellar-Mass Black Holes: These form when a massive star explodes in a supernova and collapses.

- Supermassive Black Holes: Found at the heart of galaxies, these can be billions of times the mass of our Sun.

- Intermediate-Mass Black Holes: These are a mysterious middle ground, with evidence still emerging.

__Do Black Holes Destroy Everything?__

Not necessarily! While they do swallow things, black holes don’t move through space like cosmic vacuum cleaners. In fact, they can influence their surroundings in fascinating ways—some even help form galaxies!

Scientists are still studying black holes, and discoveries about them could unlock deeper secrets about the universe, time travel, and even the nature of reality itself.

## The Sachdev-Ye-Kitaev (SYK) model

__The Sachdev-Ye-Kitaev (SYK) model__ is a fascinating concept in physics that helps scientists understand quantum chaos—how tiny particles behave in unpredictable ways.

Imagine a wild dance party where everyone is moving randomly, bumping into each other in unexpected ways. The SYK model is like that, but instead of dancers, it describes how particles interact in a chaotic manner. Scientists use this model to study extreme conditions, like those found near black holes, because it simplifies complex quantum interactions.

One of the biggest challenges in studying the SYK model is recreating these chaotic interactions in a lab. Researchers have come up with clever ways to simulate them, using tiny particles trapped in special setups that mimic the randomness needed for the model to work.

The Sachdev-Ye-Kitaev (SYK) model has deep connections to both black holes and quantum computing, making it a fascinating bridge between different areas of physics.

__Connection to Black Holes__

The SYK model helps scientists understand quantum gravity, which is crucial for studying black holes. Black holes are governed by general relativity, but their interiors involve quantum mechanics, leading to paradoxes like the information paradox—what happens to information that falls into a black hole? The SYK model provides insights into this by mimicking the chaotic behavior of black holes at a quantum level.

One key similarity is that both black holes and the SYK model exhibit maximal quantum chaos, meaning their behavior is as unpredictable as possible under quantum mechanics. This chaos is linked to the way information spreads inside a black hole, helping physicists explore how black holes might store and process information.

__Connection to Quantum Computing__

Quantum computers rely on entanglement and complex interactions between particles, which are also central to the SYK model. The randomness and strong interactions in the SYK model resemble the way quantum computers manipulate information. Some researchers believe that studying the SYK model could lead to new ways of designing quantum algorithms and understanding quantum error correction, which is essential for building stable quantum computers.

Additionally, the SYK model has been used to explore non-Fermi liquids, a type of quantum matter that behaves differently from conventional particles. Understanding these exotic states of matter could lead to breakthroughs in quantum computing hardware.

__Why This Matters__

By studying the SYK model, scientists are uncovering new ways to describe quantum gravity, improve quantum computing, and even solve long-standing mysteries about black holes. It’s an exciting area of research that could reshape our understanding of the universe!

# In this notebook we explore the SYK model

The SYK model consists of randomly interacting Majorana fermions and exhibits maximal quantum chaos, meaning it saturates the bound on quantum Lyapunov exponents, a measure of how quickly small perturbations grow over time. This behavior is closely related to the out-of-time-order correlators (OTOCs), which quantify how information spreads in a chaotic system [1],[2].

In the context of black holes, the SYK model serves as a toy model for understanding aspects of quantum gravity and holography. It shares similarities with the near-horizon dynamics of extremal black holes, particularly in how information is scrambled and lost to an observer outside the event horizon [2] [3]. The model's chaotic behavior mimics the way black holes process and obscure information, reinforcing ideas from the AdS/CFT correspondence—a theoretical framework linking quantum gravity and quantum field theory.

- get_transpiled_circuit

```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Operator
from qiskit.extensions import UnitaryGate

def get_transpiled_circuit(data,n):
    unitary_matrix=data
    # Create a UnitaryGate from the matrix
    unitary_gate = UnitaryGate(Operator(unitary_matrix))

    # Create a QuantumCircuit and add the unitary gate
    qc = QuantumCircuit(n)  # Adjust the number of qubits as needed
    qc.append(unitary_gate, range(n))

    # Transpile the circuit to decompose the unitary into standard gates
    transpiled_circuit = transpile(qc, basis_gates=['u3', 'cx'], optimization_level=3)

    # Print the transpiled circuit
    print("transpiled circuit depth:",transpiled_circuit.depth())
    print("Transpiled Circuit Display:")
    print(transpiled_circuit)
    return transpiled_circuit
```

```python
import numpy as np
from itertools import combinations
from qiskit import QuantumCircuit, Aer, execute
from qiskit.quantum_info import Statevector, entropy, SparsePauliOp, partial_trace
from openfermion import MajoranaOperator, jordan_wigner
from qiskit.circuit.library import PauliEvolutionGate
from qiskit.synthesis import SuzukiTrotter
```

```python
# Parameters
N_majorana = 6  # 6 Majorana fermions (3 qubits)
J_strength = 1000.0  # Coupling strength
t = 4 * np.pi    # Evolution time
trotter_steps = 10 # Trotter steps for accuracy
```

```python
# Generate random SYK couplings (Gaussian distribution)
terms = list(combinations(range(N_majorana), 4))
J = {}
for term in terms:
    J_val = np.random.normal(scale=J_strength * np.sqrt(6) / (N_majorana**1.5))
    print("J_val:",J_val,"for term:",term)
    J[term] = J_val

# Build SYK Hamiltonian
H = MajoranaOperator()
for indices, J_val in J.items():
    H += MajoranaOperator(
        indices,
        coefficient=J_val
    )
```

```python
print("H:",H)
print("H type:",type(H))
```

```python
# Convert to qubit Hamiltonian via Jordan-Wigner
H_qubit = jordan_wigner(H)

# Convert QubitOperator to SparsePauliOp
pauli_terms = []
for term, coeff in H_qubit.terms.items():
    pauli_str = ['I'] * 3
    for (index, op) in term:
        pauli_str[index] = op
    pauli_terms.append((''.join(pauli_str), coeff))

print("Number of Hamiltonian terms:",len(pauli_terms))
#print("Sample terms:",pauli_terms[:3])
print("All terms:",pauli_terms)

# Handle empty Hamiltonian
if pauli_terms:
    sparse_pauli_op = SparsePauliOp.from_list(pauli_terms).simplify()
else:
    sparse_pauli_op = SparsePauliOp(['III'], coeffs=[0])
print("sparse_pauli_op:",sparse_pauli_op)
```

```python
# Initialize quantum circuit
qc = QuantumCircuit(3)
qc.h([0, 1, 2])  # Prepare |+++> initial state
```

```python
# Compute entanglement entropy before scambling
state = Statevector(qc)
print(state)
rho = partial_trace(state,[1,2])
entanglement_entropy = entropy(rho, base=2)
print(f"Entanglement entropy: {entanglement_entropy:.4f}")
```

```python
# Apply evolution with custom labels
if sparse_pauli_op.size > 0 and not np.allclose(sparse_pauli_op.coeffs, 0):
    # Extract terms for visualization
    #terms_to_show = 3
    #pauli_list = sparse_pauli_op.paulis[:terms_to_show]
    #coeffs = sparse_pauli_op.coeffs[:terms_to_show]
    pauli_list = sparse_pauli_op.paulis
    coeffs = sparse_pauli_op.coeffs
    
    # Generate custom label
    label = "H = "
    for pauli, coeff in zip(pauli_list, coeffs):
        label += f"{coeff.real:.2f} {pauli.to_label()} +"
    #label = label[:-3] + "..."  # Trim trailing "+"
    label = label[:-1] #drop last +

    #evolution_gate = PauliEvolutionGate(
    #    sparse_pauli_op,
    #    time=t,
    #    label=label,  # Custom label with coefficients
    #    synthesis=SuzukiTrotter(order=2, reps=trotter_steps)
    #)

    evolution_gate = PauliEvolutionGate(
        sparse_pauli_op,
        time=t,
        label=label,  # Custom label with coefficients
    #    synthesis=SuzukiTrotter(order=2, reps=trotter_steps) # this step does not produce a unitary circuit!
    )

    u_pauli = Operator(evolution_gate)
    for i in range(trotter_steps-1):
        u_pauli = Operator(evolution_gate) @ u_pauli
    
    transc = get_transpiled_circuit(u_pauli.data,3)

    #unitary_gate = UnitaryGate(u_pauli)
    
    #build the evolution gate
    #Y = SparsePauliOp("Y")
    #X = SparsePauliOp("X")
    #Z = SparsePauliOp("Z")
    #I = SparsePauliOp("I")

    #op_list = [(Z ^ Z ^ I),(Z ^ Y ^ X),(Z ^ Y ^ Y),(Z ^ X ^ X),(Z ^ X ^ Y),( Z ^ I ^Z),(Y ^ I ^ X),(Y ^ I ^Y),(Y ^ X ^ Z),(Y ^ Y ^ Z),(X ^ I ^ X),(X ^ I ^ Y),(X ^ X ^ Z),(X ^ Y ^ Z),(I ^ Z ^ Z)]
    #operator = round(coeffs[0].real,2) * op_list[0]
    #for i in range (1, len(op_list)):
    #    operator += round(coeffs[i].real,2) * op_list[i]
    #print(operator)

    #evolution_gate = PauliEvolutionGate(
    #    operator,
    #    time=t,
    #    label=label  # Custom label with coefficients
    #)
    
    #qc.append(evolution_gate, range(3))
    #qc.append(unitary_gate, range(3))
    qc.compose(transc, [0, 1, 2], inplace=True)
    print(f"Evolution label: {label}")
else:
    print("No evolution applied.")
```

```python
# Compute entanglement entropy
state = Statevector(qc)
rho = partial_trace(state,[1,2])
entanglement_entropy = entropy(rho, base=2)
print(f"Entanglement entropy: {entanglement_entropy:.4f}")
```

```python
def get_transpiled_circuit_custom(data,n):
    unitary_matrix=data
    # Create a UnitaryGate from the matrix
    unitary_gate = UnitaryGate(Operator(unitary_matrix))

    # Create a QuantumCircuit and add the unitary gate
    qc = QuantumCircuit(n)  # Adjust the number of qubits as needed
    qc.append(unitary_gate, range(n))

    # Transpile the circuit to decompose the unitary into standard gates
    transpiled_circuit = transpile(qc, basis_gates=['u3', 'cx'], optimization_level=3)

    # Print the transpiled circuit
    #print("transpiled circuit depth:",transpiled_circuit.depth())
    #print("Transpiled Circuit Display:")
    #print(transpiled_circuit)
    return transpiled_circuit

# Parameters
N_majorana = 6  # 6 Majorana fermions (3 qubits)
J_strength = 10000.0  # Coupling strength
t = 8 * np.pi    # Evolution time
trotter_steps = 20 # Trotter steps for accuracy

def syk6_scramble():
    # Generate random SYK couplings (Gaussian distribution)
    terms = list(combinations(range(N_majorana), 4))
    J = {}
    for term in terms:
        J_val = np.random.normal(scale=J_strength * np.sqrt(6) / (N_majorana**1.5))
        J[term] = J_val

    # Build SYK Hamiltonian
    H = MajoranaOperator()
    for indices, J_val in J.items():
        H += MajoranaOperator(
        indices,
        coefficient=J_val
        )   
    # Convert to qubit Hamiltonian via Jordan-Wigner
    H_qubit = jordan_wigner(H)

    # Convert QubitOperator to SparsePauliOp
    pauli_terms = []
    for term, coeff in H_qubit.terms.items():
        pauli_str = ['I'] * 3
        for (index, op) in term:
            pauli_str[index] = op
        pauli_terms.append((''.join(pauli_str), coeff))

    # Handle empty Hamiltonian
    if pauli_terms:
        sparse_pauli_op = SparsePauliOp.from_list(pauli_terms).simplify()
    else:
        sparse_pauli_op = SparsePauliOp(['III'], coeffs=[0])

    # Initialize quantum circuit
    qc = QuantumCircuit(3)
    qc.h([0, 1, 2])  # Prepare |+++> initial state
   
    # Apply evolution with custom labels
    if sparse_pauli_op.size > 0 and not np.allclose(sparse_pauli_op.coeffs, 0):
        pauli_list = sparse_pauli_op.paulis
        coeffs = sparse_pauli_op.coeffs
    
        # Generate custom label
        label = "H = "
        for pauli, coeff in zip(pauli_list, coeffs):
            label += f"{coeff.real:.2f} {pauli.to_label()} +"
        label[:-1] #drop last +
       
        evolution_gate = PauliEvolutionGate(
          sparse_pauli_op,
          time=t,
          label=label,  # Custom label with coefficients
        #    synthesis=SuzukiTrotter(order=2, reps=trotter_steps) # this step does not produce a unitary circuit!
        )

        u_pauli = Operator(evolution_gate)
        for i in range(trotter_steps-1):
            u_pauli = Operator(evolution_gate) @ u_pauli
    
        transc = get_transpiled_circuit_custom(u_pauli.data,3)
        qc.compose(transc, [0, 1, 2], inplace=True)
    else:
        print("No evolution applied.")

    # Compute entanglement entropy
    state = Statevector(qc)
    rho = partial_trace(state,[1,2])
    entanglement_entropy = entropy(rho, base=2)
    #print(f"Entanglement entropy: {entanglement_entropy:.4f}")
    return entanglement_entropy
```

```python
ent_list=[]
for i in range(400):
    ent_list.append(round(syk6_scramble(),4))
```
